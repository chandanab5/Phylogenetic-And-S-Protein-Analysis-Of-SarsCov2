```python
file = open('pro.txt')
proteins = file.readlines()
proteinseq=[]
for pr in proteins:
    pr = pr.replace('\n', '')
    proteinseq.append(pr)
protein=[]
protein.append(''.join(proteinseq[17:32]))
protein.append(''.join(proteinseq[1:16]))
protein.append(''.join(proteinseq[33:48]))
protein.append(''.join(proteinseq[49:64]))
protein.append(''.join(proteinseq[65:94]))
covid=['Alpha','Beta','Gamma','Delta','Omicron']
colors=['forestgreen','steelblue','darkorchid','firebrick','tan']
i=0
for p in protein:
  from collections import Counter
  res = Counter(p)
  print("The S protein analysis of", covid[i], "variant:")
  i=i+1
  print ("Count of all amino acids is :\n "+ str(res))
  s = sum(res.values())
  print("Percentage of Amino acids are :\n")
  for k, v in res.items():
      pct = v * 100.0 / s
      print(k, pct)
  from matplotlib import pyplot as plt
  from matplotlib.ticker import PercentFormatter
  for f in res:
      res[f] = int(res[f])
  total = sum(res.values())
  plt.bar(res.keys(), [v/total for v in res.values()], color=colors[i-1])
  plt.gca().yaxis.set_major_formatter(PercentFormatter(xmax=1, decimals=0))
  plt.grid(axis='y')
  plt.title('Spike Protein Analysis')
  plt.xlabel('Amino acids')
  plt.ylabel('Percentage')
  plt.show()
```
